```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("0000000000002747_training_twitter_x_y_train.csv")
test=pd.read_csv("0000000000002747_test_twitter_x_test.csv")
```

```python
train.isnull().sum()
```

```python
categories=train["airline_sentiment"]
```

```python
category=[]
for i in categories:
    category.append(i)
```

```python
category
```

```python
word=train["text"]
```

```python
words=[]
for i in word:
    words.append(i)
```

```python
words
```

```python
from nltk.corpus import stopwords
import string
stop=set(stopwords.words('english'))
punctuations=list(string.punctuation)
stop.update(punctuations)
stop.update(" ")
```

```python
from nltk.tokenize import word_tokenize
text = [] 
for i in range(len(words)):
    text.append([]) 
    for w in word_tokenize(words[i]):
        if w not in stop:
            text[i].append(w)
```

```python
text
```

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
from nltk.corpus import wordnet
from nltk import pos_tag
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
documents = []
for i in range(len(text)):
    documents.append([])
    for word in text[i]:
        clean_word = lemma.lemmatize(word, pos = get_simple_pos(pos_tag(list(word))[0][1]))
        documents[i].append(clean_word.lower())
```

```python
documents[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text_documents=[" ".join(document) for document in documents]
```

```python
text_documents
```

```python
count_vec=CountVectorizer(max_features=2000,ngram_range=(1,2))
x_train_features=count_vec.fit_transform(text_documents)
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```
