```python
import numpy as np
import pandas as pd
```

```python
training_data=np.genfromtxt("0000000000002417_training_boston_x_y_train.csv",delimiter=",")
testing_data=np.genfromtxt("0000000000002417_test_boston_x_test.csv",delimiter=",")
```

```python
training_data.shape,testing_data.shape
```

```python
x=training_data[:,0:13]
y=training_data[:,13]
x.shape,y.shape
```

```python
def step_gd(x,y,alpha,m,M,N):
    m_slope=np.zeros(M)
    for j in range(N):
        for i in range(M):
            m_slope[j] = m_slope[j] + (2/M) * (y[i]-m[j]*x[i,j]) * x[i,j]
            
    for j in range(N):
        m[j]=m[j]-alpha*m_slope[j]
    
    return m
```

```python
def gd(x,y,alpha,num_iter):
    M=x.shape[0] #no.of samples
    N=x.shape[1] #no. of features
    m=np.zeros(N)
    
    cost=np.zeros(num_iter)
    #gradient descent algo
    for i in range(num_iter):
        m=step_gd(x,y,alpha,m,M,N)
        cost=total_cost(x,y,M,N)
    return cost, m
```

```python
def total_cost(x,y,M,N):
    cost=np.zeros(M)
    for j in range(N):
        for i in range(M):
            cost[j] = cost[j] + (1/M) * ((y[i]-m[j]*x[i,j]) ** 2)
        print(j,"Cost: ", cost[j])
    return cost
```

```python
def predict(x,m):
    return x*m
```

```python
cost,m= gd(x,y,0.00001,1000)
print(m)
```
